* Le blanc est la représentation de la monarchie française.

La proportion de ce drapeau est 2:3, ce qui signifie que la hauteur mesure 2/3 de la largeur

## 2) Création d'un programme python pour créer une image du drapeau français
On se propose ici d'utiliser les librairies pillow et matplotlib déjà utilisées dans le TD n°5 pour créer et sauvegarder une image du drapeau français en respectant les consignes ci-dessus.

```python
# -*- coding: utf-8 -*-

import PIL.Image as pil
import matplotlib.pyplot as plt

""" 1) Drapeau français """

# Taille de l'image : L et H sont la largeur et la hauteur de l'image en pixels
L = 300
H = 200
print("definition de l'image : ", L, H)

# Création de l'image
img_out = pil.new("RGB", [L,H])
# parcouurs des lignes
for j in range(H):
    # parcours des colonnes
    for i in range(L):
        if i < L//3 :
            img_out.putpixel((i,j), (5,20,64))
        elif i < (2*L)//3 :
            img_out.putpixel((i,j), (255,255,255))
        else :
            img_out.putpixel((i,j), (236,25,32))

# sauvegarde de l'image crée dans le fichier butterfly_rouge.jpg
img_out.save("drapeau_francais.jpg")

# affichage de l'image initiale
plt.imshow(img_out)
plt.title("drapeau_francais.jpg")
plt.show()

# fermeture du fichier créé
img_out.close()
```

# II) Application à d'autres drapeaux
On se propose ici d'adapter le code du drapeaux français pour créer d'autres drapeaux, notamment le drapeau allemand.
## 1) Le drapeau allemand
En vous aidant du site suivant :
<p style="text-align:center;"><a href="https://fr.wikipedia.org/wiki/Drapeau_de_l%27Allemagne#Conception">https://fr.wikipedia.org/wiki/Drapeau_de_l%27Allemagne#Conception</a></p>

Modifier le code du code suivant pour créer le drapeau allemand et l'enregistrer dans le fichier "drapeau_allemand.jpg".

```python
# -*- coding: utf-8 -*-

import PIL.Image as pil
import matplotlib.pyplot as plt

""" 1) Drapeau français """

# Taille de l'image : L et H sont la largeur et la hauteur de l'image en pixels
L = 300
H = 200
print("definition de l'image : ", L, H)

# Création de l'image
img_out = pil.new("RGB", [L,H])
# parcours des colonnes
for i in range(L):
    # parcouurs des lignes
    for j in range(H):
        if j < H//3 :
            img_out.putpixel((i,j), (0,0,0))
        elif j < (2*H)//3 :
            img_out.putpixel((i,j), (255,0,0))
        else :
            img_out.putpixel((i,j), (255,204,0))

# sauvegarde de l'image crée dans le fichier butterfly_rouge.jpg
img_out.save("drapeau_allemand.jpg")

# affichage de l'image initiale
plt.imshow(img_out)
plt.title("drapeau_allemand.jpg")
plt.show()

# fermeture du fichier créé
img_out.close()
```

## 2) Le drapeau japonais
En vous aidant du site suivant :
<p style="text-align:center;"><a href="https://fr.wikipedia.org/wiki/Drapeau_du_Japon">https://fr.wikipedia.org/wiki/Drapeau_du_Japon</a></p>

Modifier le code suivant pour créer le drapeau japonais et l'enregistrer dans le fichier "drapeau_japon.jpg". Pour tester si un pixel appartient au disque central, il faut comparer la distance entre le pixel(i,j) et le centre du drapeau (150,100). il faut donc comparer (i-150)^2 + (j-100)^2 avec le rayon du disque. En python pour calculer le carré d'un nombre, on utilise deux étoiles. Par exemple <code>3**2 = 9</code>.

```python
# -*- coding: utf-8 -*-

import PIL.Image as pil
import matplotlib.pyplot as plt

# Taille de l'image : L et H sont la largeur et la hauteur de l'image en pixels
L = 300
H = 200
print("definition de l'image : ", L, H)
x_centre = 150
y_centre = 100

# Création de l'image
img_out = pil.new("RGB", [L,H])
# parcours des colonnes
for i in range(L):
    # parcouurs des lignes
    for j in range(H):
        if (i-x_centre)**2 + (j-y_centre)**2 < ((H*3)//10)**2 :
            img_out.putpixel((i,j), (188,0,45))
        else :
            img_out.putpixel((i,j), (255,255,255))

# sauvegarde de l'image crée dans le fichier butterfly_rouge.jpg
img_out.save("drapeau_japon.jpg")

# affichage de l'image initiale
plt.imshow(img_out)
plt.title("drapeau_japon.jpg")
plt.show()

# fermeture du fichier créé
img_out.close()
```
